# Assigning classes to words

```python
df.drop([230,231,232], axis = 0,inplace=True)
```

```python
def buc(x):
    if 'ZOMATO' in x:
        return 'Online Food'
    if 'SWIGGY' in x:
        return 'Online Food'
    if 'ZERODHA' in x:
        return 'Broker'
    if 'RECHARG' in x:
        return 'Recharge and Subcription'
    if x[-1]=='UPI':
        return 'Utility'
    if 'NEFT' in x:
        return 'Funds Transfer'
    if 'AMZN' in x:
        return 'Online Shoping'
    if 'FLIPKART' in x:
        return 'Online Shoping'
    if 'CREDIT' in x:
        return 'CREDIT'
    
        
df['bucket'] = df['text_token'].apply(buc)
```
